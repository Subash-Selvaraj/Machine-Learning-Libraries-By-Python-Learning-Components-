```python
##indexing columns and training delimitres
data = ('index,a,b,c\n'
       '4,apple,bat,5.7\n'
       '8,orange,cow,10')
```

```python
df = pd.read_csv(StringIO(data))
df
```

```python
#change the first column as the index in the dataframe(by using index_col funtion)
df = pd.read_csv(StringIO(data),index_col=0)
df
```

```python
data = ('a,b,c\n'
           '4,apple,bat,5.7\n'
           '8,orange,cow,10') #while create only three columns in csv there will be the filrst column is considerd as a index's  
```

```python
df=pd.read_csv(StringIO(data))
df
```

```python
#to avoid the automatic indexing by using index_col = fasle
df=pd.read_csv(StringIO(data),index_col=False)
df
```

```python
#we can also combine usecols and index_col
data = ('a,b,c\n'
           '4,apple,bat,5.7\n'
           '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data),usecols=['b','c'],index_col=False)
```
