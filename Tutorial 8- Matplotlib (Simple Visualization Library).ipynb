# MatplotLip Tutorial

###### Matplotlib is a plotting library for the python programming language and its numerical mathematics extension Numpy.it provides an object-oriented API for embedding plots into applications using general-GUI toolkits like Tkinter wxPython,QT, or GTK+

###### Some of the majot Pros of Matplotlib are:

 - Generally easy to get started for simple plots
 - Support for custom lables and txts
 - Great control of every element in a figure
 - High-quality output in many formats
 - Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
## simple examples

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
#let's display the daigram
#plt scatter
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 20')
plt.savefig('Test.png')
```

```python
#plt plot #in plot we will get staright line
plt.plot(x,y)
```

```python
#if i square the x and assign in to y then it will 
y = x*x
plt.plot(x,y,'r*', linestyle = 'dashed',linewidth=2, markersize=12) # there is lot of option in this plot jusy click shift tab
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
#creat subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'k*')
plt.subplot(2,2,3)
plt.plot(x,y,'y')
plt.subplot(2,2,4)
plt.plot(x,y,'c')
```

```python
# give complicated array and select the value and make a plot
x = np.arange(1,11)
y = 3 * x / 5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
```

```python
# compute the x and y coordinates for points on a sine curve
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("Sign wave form")

#plot the points using matplotlib
plt.plot(x,y)
```

```python
#subplot()
#compute the x and y coordinates for the points on sine and cosine curves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#set up a subplot grid that has height 2 widht 1,
#and set the first such subplot as active
plt.subplot(2,1,1)

#make the first plot
plt.plot(x,y_sin,'--g*')
plt.title("sine")
#set the second subplot as an active, and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'*c--')
plt.title("cosine")
```

### Bar Graph

```python
## Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y, color = 'r')
plt.bar(x2,y2, color = 'g')
plt.title("Bar graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## HIstograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,4,79,31,27])
plt.hist(a)
plt.title("Histogram")
```

### Box Plot using Matplotlib

```python
data = [np.random.normal(0, std,100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data, vert = True,patch_artist=True #color)
plt.title("Box Plot of Percentile")
```

### Pie Chart

```python
#Data to plot the pie chart

labels = 'Python','C++','java','Ruby'
sizes = [215,130,245,210]
colors = ['gold','yellow','lightcoral','lightskyblue']
explode = (0.1,0,0,0)

#plot
plt.pie(sizes, explode=explode,colors=colors,labels=labels)
plt.title('peichart')
```
